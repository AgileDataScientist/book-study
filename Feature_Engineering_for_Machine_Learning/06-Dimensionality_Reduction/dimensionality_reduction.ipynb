```python
# Imports

## Plotting imports
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

## ML imports
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
# Loading the data

digits_data = datasets.load_digits()
labels = digits_data.target
```

```python
# Viewing the shape

digits_data.images.shape
```

```python
# Plotting the image

plt.imshow(digits_data.images[0])
plt.axis('off');
```

```python
# Getting the number of images

n = len(digits_data.images)
```

```python
# Reshaping the data

image_data = digits_data.images.reshape((n, -1))
image_data.shape
```

```python
# Fitting and transforming using PCA

pca_transformer = PCA(n_components=0.8)
pca_images = pca_transformer.fit_transform(image_data)
```

```python
# Viewing the explained variance of the components

pca_transformer.explained_variance_ratio_
```

```python
# Viewing the sum of the explained variance of the first 3 components

pca_transformer.explained_variance_ratio_[:3].sum()
```

```python
# Plotting the results

fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection='3d')
for i in range(100):
    ax.scatter(
        pca_images[i,0],
        pca_images[i,1],
        pca_images[i,2],
        marker=r'${}$'.format(labels[i]),
        s=64
    )
ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
ax.set_zlabel('PC 3');
```
